```python
import pyarma as pa
import numpy as np
import matplotlib.pyplot as plt

def plot_quantity(shared_filename, quantity, plotname):
    
    #IMPORT THE RESULTS

    data1 = pa.mat(); data1.load(shared_filename + '40.csv') 
    data2 = pa.mat(); data2.load(shared_filename + '60.csv') 
    data3 = pa.mat(); data3.load(shared_filename + '80.csv') 
    data4 = pa.mat(); data4.load(shared_filename + '100.csv') 

    #ARRANGE LISTS FOR PLOTTING

    Tdata = []; Qdata1 = []; Qdata2 = []; Qdata3 = []; Qdata4 = [];
    for i in range(len(data1)):
        if i < round(len(data1)/2):
            Tdata.append(data1[i])
        else:
            Qdata1.append(data1[i])
            Qdata2.append(data2[i])
            Qdata3.append(data3[i])
            Qdata4.append(data4[i])

    #PLOT THE RESULTS
        
    plt.figure(figsize=(9, 6))
    plt.rcParams.update({'font.size': 20})
    linestyle='dashed'

    plt.plot(Tdata,Qdata1, label='L=40', linestyle=linestyle, linewidth=1.5, color='cornflowerblue')
    plt.plot(Tdata, Qdata2, label='L=60', linestyle=linestyle, linewidth=1.5, color='royalblue')
    plt.plot(Tdata,Qdata3, label='L=80', linestyle=linestyle, linewidth=1.5, color='midnightblue')
    plt.plot(Tdata,Qdata4, label='L=100', linestyle=linestyle, linewidth=1.5, color='black')

    plt.legend(); plt.xlabel('T$[J/k_B]$'); plt.ylabel(quantity)
    plt.savefig(plotname)

#PLOTS
plot_quantity('energy', 'Average energy $[J]$', 'MC500000_average_energy.pdf')
plot_quantity('magnetism', 'Average Magnetization', 'MC500000_finescan_average_magnetization.pdf')
plot_quantity('susceptibility', 'Magnetic Susceptibility [$1/J$]', 'MC500000_finescan_magnetic_susceptibility.pdf')
plot_quantity('specificheat', 'Specific Heat Capacity $[k_B]$', 'MC500000_finescan_specific_heat_capacity.pdf')
```
