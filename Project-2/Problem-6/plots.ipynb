```python
import pyarma as pa
import numpy as np
import matplotlib.pyplot as plt
```

```python
eigvec = pa.mat() 
eigvec.load('jacobi_eigenvectors.txt') 

eigval = pa.mat() 
eigval.load('jacobi_eigenvalues.txt') 

x_hat = np.linspace(0,1,pa.size(eigval)[0] + 1)

def eigv_analytical(a, b, N):
    A = np.zeros(a)
    for i in range(a):
        A[i] = np.sin(((i+1)*b*np.pi)/(N + 1))
    return A
```

```python
for i in range(3):
    plt.plot(x_hat[1:], eigvec[:,i], label=i+1)
    plt.legend(); plt.xlabel('x_hat'); plt.ylabel('f')
plt.savefig('disc.pdf')
```

```python
for i in range(3):
    plt.plot(x_hat[1:], eigv_analytical(pa.size(eigval)[0], i+1,pa.size(eigval)[0]), '--', label=i+1)
    plt.legend(); plt.xlabel('x_hat'); plt.ylabel('f')
plt.savefig('analyt.pdf')
```
